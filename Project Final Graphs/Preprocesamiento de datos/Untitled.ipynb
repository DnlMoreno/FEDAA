```python
import pandas as pd
```

```python
result_tendencias = pd.read_csv("Tendencias_politicas_usuarios.csv", sep = ";")
result_componentes = pd.read_csv("Componentes_fuertemente_conexas_filtradas.csv", sep = ";")
```

```python
result_tendencias
```

```python
result_componentes
```

```python
result_componentes = result_componentes.fillna(0)
```

```python
lista = []
for i in range(len(result_componentes)):
    for j in range(1737):
        if result_componentes.iloc[i,j] != 0:
            usuario = result_tendencias[result_tendencias['Usuario'] == result_componentes.iloc[i,j]]
            
            #componentes_concat = pd.concat([followee, follower], axis=0)
            lista.append(usuario)
```

```python
componentes_tendencias = pd.concat(lista)
componentes_tendencias
```

```python
componentes_tendencias.to_csv("componentes_con_tendencias.csv", index = False, sep = ';')
```
