```python
import pandas as pd
```

```python
# Datos
twitter_users = pd.read_csv("twitter_users.csv", sep = ";")
twitter_connections = pd.read_csv("twitter_connections.csv", sep = ";")
```

```python
twitter_users
```

```python
for i in range(len(twitter_users)):
    twitter_users['user_id'].iloc[i] = i
twitter_users
```

```python
twitter_users_tweets = twitter_users.iloc[:, 1:5]
twitter_users_tweets = twitter_users_tweets.drop(['number_tweets'], axis=1)
twitter_users_tweets = twitter_users_tweets.rename(columns={'friends_count': 'followee_count'})
twitter_users_tweets
```

```python
twitter_users = twitter_users.iloc[:, 0:2]
twitter_users
```

```python
twitter_users = twitter_users.rename(columns={'user_id': 'id_followee'})
twitter_users
```

```python
temp = twitter_users[:]
temp = temp.rename(columns={'id_followee': 'id_follower'})
temp
```

```python
twitter_connections
```

```python
twitter_connections_update_v1 = pd.merge(twitter_connections, twitter_users, left_on='followee', right_on='user_name')
```

```python
twitter_connections_update_v1 = twitter_connections_update_v1.iloc[:, 0:3]
largo_total = len(twitter_connections_update_v1)
twitter_connections_update_v1
```

```python
twitter_connections_update_v2 = pd.merge(twitter_connections_update_v1, temp, left_on='follower', right_on='user_name')
```

```python
twitter_connections_update_v2 = twitter_connections_update_v2.iloc[:,0:4]
```

```python
twitter_connections_update_v2
```

```python
twitter_connections_update_v2 = twitter_connections_update_v2.drop(twitter_connections_update_v2[twitter_connections_update_v2['follower']=='Cooperativa'].index)
a = len(twitter_connections_update_v2)
print(f"Cantidad de valores eliminados de Cooperativa: {largo_total - a}")

twitter_connections_update_v2 = twitter_connections_update_v2.drop(twitter_connections_update_v2[twitter_connections_update_v2['follower']=='soyvaldiviacl'].index)
b = len(twitter_connections_update_v2)
print(f"Cantidad de valores eliminados de soyvaldiviacl: {a-b}")

twitter_connections_update_v2 = twitter_connections_update_v2.drop(twitter_connections_update_v2[twitter_connections_update_v2['follower']=='latercera'].index)
c = len(twitter_connections_update_v2)
print(f"Cantidad de valores eliminados de latercera: {b-c}")

twitter_connections_update_v2 = twitter_connections_update_v2.drop(twitter_connections_update_v2[twitter_connections_update_v2['follower']=='elmostrador'].index)
d = len(twitter_connections_update_v2)
print(f"Cantidad de valores eliminados de elmostrador: {c-d}")
```

```python
twitter_connections_update_v2_tweets = pd.merge(twitter_connections_update_v2, twitter_users_tweets, left_on='follower', right_on='user_name')
twitter_connections_update_v2_tweets
```

```python
twitter_connections_update_v2_tweets = twitter_connections_update_v2_tweets.drop(['user_name'], axis=1)
twitter_connections_update_v2_tweets
```

```python
set_1 = set(twitter_connections_update_v2_tweets['followee'])
len(set_1)
```

```python
set_2 = set(twitter_connections_update_v2_tweets['follower'])
len(set_2)
```

```python
twitter_connections_update_v2_tweets.to_csv("twitter_connections_update.csv", index = False, sep = ';')
```
