```python
twitter_connections_update = twitter_connections_update.drop(twitter_connections_update[twitter_connections_update['follower']=='Cooperativa'].index)
a = len(twitter_connections_update)
print(f"Cantidad de valores eliminados de Cooperativa: {largo_total - a}")

twitter_connections_update = twitter_connections_update.drop(twitter_connections_update[twitter_connections_update['follower']=='soyvaldiviacl'].index)
b = len(twitter_connections_update)
print(f"Cantidad de valores eliminados de soyvaldiviacl: {a-b}")

twitter_connections_update = twitter_connections_update.drop(twitter_connections_update[twitter_connections_update['follower']=='latercera'].index)
c = len(twitter_connections_update)
print(f"Cantidad de valores eliminados de latercera: {b-c}")

twitter_connections_update = twitter_connections_update.drop(twitter_connections_update[twitter_connections_update['follower']=='elmostrador'].index)
d = len(twitter_connections_update)
print(f"Cantidad de valores eliminados de elmostrador: {c-d}")
```
