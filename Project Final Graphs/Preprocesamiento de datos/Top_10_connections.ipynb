```python
import pandas as pd
```

```python
df = pd.read_csv("twitter_connections_update.csv", sep = ";")

influyentes = ['pachidiaze', 'DBustosKorts', 'Christhitho', 'LaMazurquita', '24genco', 'tatto_falconi', 'jcmr2009', 'RodrigoRACT', 'FcoUlloaMontes', 'cecilialosrios']
influenciables = ['panoramasgratis', 'PatiVeloso_D24', 'barriolasanimas', 'MunicipalVald', 'ferr_altermatt', 'gealtor', 'pautalosrios', 'Emol', 'Shilevision_CL', 'patriciorosas']
```

```python
df_v = df.iloc[:, 0:2]
df_v
```

```python
df_v[df_v['followee'] == 'pachidiaze']
```

```python
df_influenciables = df_v[df_v['follower'].isin(influenciables)]
```

```python
df_influenciables
```

```python
df_influencers = df_v[df_v['followee'].isin(influyentes)]
```

```python
df_influencers
```
