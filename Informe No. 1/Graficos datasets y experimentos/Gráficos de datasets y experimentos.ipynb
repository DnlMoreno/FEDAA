```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
import numpy as np
from sympy import *
```

```python
def recortar(n, lista, aux):
    for j in range(n):
        lista.append(aux[j])
        
    return lista
```

## Lectura de datos

```python
data1 = open("C:/Users/gamor/Documents/GitHub/Fundamentos-de-Estructuras-de-Datos-y-Algoritmos/Informe No. 1/Generación de datasets/Entradas algoritmos de ordenamiento/Entrada datos casiOrdenados.txt", "r")
data2 = open("C:/Users/gamor/Documents/GitHub/Fundamentos-de-Estructuras-de-Datos-y-Algoritmos/Informe No. 1/Generación de datasets/Entradas algoritmos de ordenamiento/Entrada datos mitadOrdenados.txt", "r")
data3 = open("C:/Users/gamor/Documents/GitHub/Fundamentos-de-Estructuras-de-Datos-y-Algoritmos/Informe No. 1/Generación de datasets/Entradas algoritmos de ordenamiento/Entrada datos aleatorios.txt", "r")
data4 = open("C:/Users/gamor/Documents/GitHub/Fundamentos-de-Estructuras-de-Datos-y-Algoritmos/Informe No. 1/Generación de datasets/Entradas algoritmos de ordenamiento/Entrada datos ordenadosReversa.txt", "r")
data5 = open("C:/Users/gamor/Documents/GitHub/Fundamentos-de-Estructuras-de-Datos-y-Algoritmos/Informe No. 1/Generación de datasets/Entradas algoritmos de ordenamiento/Entrada datos valoresRepetidos.txt", "r")
lista_data = [data1, data2, data3, data4, data5]
lista_data1 = []
lista_data2 = []
lista_data3 = []
lista_data4 = []
lista_data5 = []

n = 10000 # Tamaño a graficar

# Recorro todos los archivos
for datos in lista_data:
    flag = True
    for linea in datos:
        aux = linea.split()
        
        # Guarda el tamaño que aparece en el txt
        if len(aux) == 1: 
            size = aux[0]

        # Guarda las 5 instancias del primer tamaño en un lista
        if len(aux)/5 == n and flag == True:
            
            if data1 == datos:
                lista_data1 = recortar(n, lista_data1, aux)
                casiOrdenados = [int(i) for i in lista_data1]
                flag = False
                
            if data2 == datos:
                lista_data2 = recortar(n, lista_data2, aux)
                mitadOrdenados = [int(i) for i in lista_data2]
                flag = False
                
            if data3 == datos:
                lista_data3 = recortar(n, lista_data3, aux)
                aleatorios = [int(i) for i in lista_data3]
                flag = False 

            if data4 == datos:
                lista_data4 = recortar(n, lista_data4, aux)
                ordenadosReversa = [int(i) for i in lista_data4]
                flag = False

            if data5 == datos:
                lista_data5 = recortar(n, lista_data5, aux)
                valoresRepetidos = [int(i) for i in lista_data5]
                flag = False
```
